## Coefficient of Variation (CV)
---

■ The coefficient of variation (CV) is a measure of relative variability. It is the ratio of the standard deviation to the mean (average). For example, the expression “The standard deviation is 15% of the mean” is a CV.\
■ The CV is particularly useful when you want to compare results from two different surveys or tests that have different measures or values. For example, if you are comparing the results from two tests that have different scoring mechanisms. If sample A has a CV of 12% and sample B has a CV of 25%, you would say that sample B has more variation, relative to its mean. The formula for the coefficient of variation is:\
\
$$ Coefficient of Variation = (Standard Deviation/ Mean) * 100. $$ \
\

### Definition: Percentiles
■ Split the series in 100 equal parts\
■ In general the nth percentile is a value such that n% of the observations fall 
at or below or it

image.png

- First quartile (designated Q1) also called the lower quartile = the 25th percentile
- Second quartile (designed Q2) also called the Median = the 50th percentile
- Third quartile (designed Q3) also called the upper quartile = the 75th percentile

### Computation: Percentiles
The i - th percentile of a list of N ordered values (sorted from least to 
greatest) is the score computed according the following method:

1. Compute the ordinal rank by the formula
$$ R = \frac{i}{100} * (N + 1) $$

    If R is an integer then i is the score with the index R, otherwise take $I_R$
    and $F_R$ as an integer part and fractional part of R respectively;

2. Let $s_1$ is the score with index $I_R$ and $s_2$ the score with index $I_R+1;$
3. Compute $i = s_1+F_R(s_2 - s_1)$

#### Example: 25th percentile of 20 Quiz scores
---
image.png\
and therefore $I_R$ = 5, $F_R$ = 0.25, $s_1$ = 5 and $s_2$ = 5.\
**The 25th percentile is 5+0.25(5-5) = 5**

## Kurtosis
■ a measurement about the extremities (i.e. tails) of the 
distribution of data, and therefore provides an indication of the 
presence of outliers.\
■ Formula for calculus:
$$ \alpha_4 = \frac{\frac{1}{N}‎\sum_{i=1}^{N}(x_i‎‏-\overline{x})^4}{S^4} - 3$$

### The degrees of kurtosis
image.png/
- **Mesokurtic:** Distributions that are moderate in breadth and curves with a medium peaked height.
- **Leptokurtic:** More values in the distribution tails and more values close to the mean (i.e. sharply peaked with heavy tails)
- **Platykurtic:** Fewer values in the tails and fewer values close to the mean (i.e. the curve has a flat peak and has more dispersed scores with lighter tails).

| Types       | Kurtosis | Excess Kurtosis |   |   |
|-------------|----------|-----------------|---|---|
| Leptokurtic | >3       | >0              |   |   |
| Platykurtic | <3       | <0              |   |   |
| Mesokurtic  | =3       | =0              |   |   |

#### Example: Kurtosis with `scipy.stats`

```python
# Python code example
from scipy.stats import kurtosis

mu = 1
sigma = 3
noise = np.random.normal(mu, sigma, size=1000)

def kurtosis_type(var):
    if var > 3:
        return "Leptokurtic"
    if var < 3:
        return "Platykurtic"
    if var == 3:
        return "Mesokurtic"
    
print(kurtosis(noise, fisher=False, bias=False))
print(f"Degree of kurtosis: {kurtosis_type(kurtosis(noise, fisher=False))}")
```

```python
# Histogram plot
plt.hist(noise, histtype='step')
plt.show()
```

#### Example: Filter with Quantile

```python
# Filter at the 2.5 and 97.5 percentiles
cols = wine_example.columns

Q1 = wine_example[cols].quantile(0.025)
Q3 = wine_example[cols].quantile(0.975)
IQR = Q3 - Q1

df_no_outlier = wine_example[~((wine_example[cols] < (Q1 - 1.5 * IQR)) |(wine_example[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

"""
No. are the same so there's no outlier outside 
the interval formed by the 2.5 and 97.5 percentiles
"""
print(f"No.row before remove outliers: {len(wine_example)}")
print(f"No.row after remove outliers: {len(df_no_outlier)}")
```

```python
# Filter at the 25 and 75 percentiles

Q1 = wine_example[cols].quantile(0.25)
Q3 = wine_example[cols].quantile(0.75)
IQR = Q3 - Q1

df_no_outlier = wine_example[~((wine_example[cols] < (Q1 - 1.5 * IQR)) |(wine_example[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

"""
There're 17 outliers outside the interval formed
by the 25 and 75 percentiles
"""
print(f"No.row before remove outliers: {len(wine_example)}")
print(f"No.row after remove outliers: {len(df_no_outlier)}")
```

## Hampel filter
■ Outliers considered as the values outside the interval (I) formed by the median, plus or minus 3 median absolute deviations (MAD). \
■ The values of the lower and upper percentiles (and thus the lower and upper limits of the interval) can be computed with the quantile() function.

$$I=[median−3⋅MAD;median+3⋅MAD]$$

where MAD is the median absolute deviation and is defined as the median of the absolute deviations from the data’s median 
$$ ( X median(X) =): MAD median(| X X |) $$



**Extra:** The Hampel filter is generally used to detect anomalies in data with a timeseries structure. It basically consists of a sliding window of a parameterizable size. For each window, each observation will be compared with the Median Absolute Deviation (MAD). The observation will be considered an outlier in the case in which it exceeds the MAD by n times (the parameter n is also parameterizable). For more details, see the following Medium post as well as the Wikipedia entry on MAD. [hampel 0.0.5](https://pypi.org/project/hampel/)

https://medium.com/wwblog/clean-up-your-time-series-data-with-a-hampel-filter-58b0bb3ebb04
https://en.wikipedia.org/wiki/Median_absolute_deviation

**Medium:** [Outlier-Detection-with-hampel-filter](https://towardsdatascience.com/outlier-detection-with-hampel-filter-85ddf523c73d)

#### Example: Hampel filter on Wine Alcohol

```python
def hampel_filter_forloop(input_series, window_size, n_sigmas=3):
    
    n = len(input_series)
    new_series = input_series.copy()
    k = 1.4826 # scale factor for Gaussian distribution
    
    indices = []
    
    # possibly use np.nanmedian 
    for i in range((window_size),(n - window_size)):
        x0 = np.median(input_series[(i - window_size):(i + window_size)])
        S0 = k * np.median(np.abs(input_series[(i - window_size):(i + window_size)] - x0))
        if (np.abs(input_series[i] - x0) > n_sigmas * S0):
            new_series[i] = x0
            indices.append(i)
    
    return new_series, indices
```

```python
rw_series = wine_example['alcohol']
res, detected_outliers = hampel_filter_forloop(rw_series, 10)

print(f"No. Cleaned: {len(res)}")
print(f"No. Outliers: {len(detected_outliers)}")
```

## Statistical tests: Grubbs’s test

■ The Grubbs test detects one outlier at a time (highest or lowest value),\
so the null and alternative hypotheses are as follows:\
\
    $H_0$: The highest value is not an outlier\
    $H_1$: The highest value is an outlier\
    \
if we want to test the highest value, or:\
\
$H_0$: The lowest value is not an outlier\
$H_1$: The lowest value is an outlier\
   \
if we want to test the lowest value.\
■ Note that the Grubbs test is not appropriate for sample size of 6 or less 
(n≤6)

■ If you suspect that the maximum value in the data set may be an outlier you can 
use the following test statistic.\
image.png\
■ If you suspect that the minimum value in the data set may be an outlier you can 
use the following test statistic\
image.png\
■ If the ***p-value*** is less than the chosen **significance threshold** (generally α=0.05) 
then the null hypothesis is rejected, we will conclude that the lowest/highest 
value is an outlier. \
■ On the contrary, if the ***p-value*** is greater or equal than the **significance level**, the 
null hypothesis is not rejected, and we will conclude that, based on the data, we 
do not reject the hypothesis that the **lowest/highest value is not an outlier.**

#### Example: Grubbs’s test on sample array

```python
import numpy as np
from outliers import smirnov_grubbs as grubbs

#define data
data = np.array([1, 5, 14, 15, 15, 14, 19, 17, 16, 20, 21, 28, 11, 9, 29, 40, 200])

#perform Grubbs' test to see if minimum value is an outlier
print(grubbs.min_test(data, alpha=.05))
#perform Grubbs' test to see if maximum value is an outlier
print(grubbs.max_test(data, alpha=.05))
```

## Statistical tests: Rosner’s test
■ Rosner’s test for outliers has the advantages that: 
- it is used to detect several outliers at once (unlike Grubbs and Dixon test which must be performed iteratively to screen for multiple outliers), and
- it is designed to avoid the problem of masking, where an outlier that is close in value to another outlier can go undetected.

■ Unlike Dixon test, note that Rosner test is most appropriate when the sample size is large (n≥20).\
■ To perform the Rosner test we use the rosnerTest() function from the {EnvStats} package.


The Generalized Extreme Studentized Deviate (ESD) test for outliers can be used to search for outliers in a univariate data set, which approximately follows a normal distribution. A description of the algorithm is, e.g., given at Nist or [Rosner1983].

#### Example: Rosner’s test on arrays

```python
from __future__ import print_function, division
from PyAstronomy import pyasl

x = np.array([float(x) for x in "-0.25 0.68 0.94 1.15 1.20 1.26 1.26 1.34 1.38 1.43 1.49 1.49 \
          1.55 1.56 1.58 1.65 1.69 1.70 1.76 1.77 1.81 1.91 1.94 1.96 \
          1.99 2.06 2.09 2.10 2.14 2.15 2.23 2.24 2.26 2.35 2.37 2.40 \
          2.47 2.54 2.62 2.64 2.90 2.92 2.92 2.93 3.21 3.26 3.30 3.59 \
          3.68 4.30 4.64 5.34 5.42 6.01".split()])

# Apply the generalized ESD
r = pyasl.generalizedESD(x, 10, 0.05, fullOutput=True)

print("Number of outliers: ", r[0])
print("Indices of outliers: ", r[1])
print("        R      Lambda")
for i in range(len(r[2])):
    print("%2d  %8.5f  %8.5f" % ((i+1), r[2][i], r[3][i]))

# Plot the "data"
plt.plot(x, 'b.')
# and mark the outliers.
for i in range(r[0]):
    plt.plot(r[1][i], x[r[1][i]], 'rp')
plt.show()
```
