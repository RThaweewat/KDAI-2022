**ตัวอย่างที่ 3** ตัวอย่าง Object-Oriented Porgramming



```
class Car(object):
  def __init__(self, model, passengers, color, speed):
    self.model = model
    self.passengers = passengers
    self.color = color
    self.speed = speed

  def accelerate(self):
    self.speed = self.speed + 2
    print (self.speed)

bmw = Car("BMW", 4, "red", 5)
ferrari = Car("Ferrari", 2, "black", 10)
ford = Car("Ford", 6, "blue", 6)

bmw.accelerate()
print (bmw.color)

ferrari.accelerate()
print (ferrari.color)
ferrari.accelerate() 
#note that the speed has been updated from the previous accelerate call

print (ford.passengers)
ford.accelerate()
```

**การเขียนฟังก์ชันเวียนเกิด (Recursive Function)**

ฟังก์ชันเวียนเกิด คือ ฟังก์ชันที่เรียกตัวเอง อาจใช้แทนการวน loop ได้ แต่มักจะใช้เมื่อเกิดการกระทำซ้ำ ๆ ในโครงสร้างแบบต้นไม้ เช่น การหา max กับ min ด้วยวิธีแบ่งแยกและเอาชนะ (Divide and Conquer) หรือการหาจำนวนฟิโบนักชีตัวที่ n เป็นต้น 

**ตัวอย่าง 11** สำหรับจำนวนจริง x, y ใด ๆจำนวนฟิโบนักชี (Fibonacci numbers) คือลำดับของจำนวนเต็มที่กำหนดโดย

F(0) = 0, F(1) = 1,     F(n) = F(n-1) + F(n-2) สำหรับจำนวนนับ n >= 2 

ดังนั้นจะได้ลำดับ   0, 1, 1, 2, 3, 5, 8, 13, ...

จงเขียนโปรแกรมรับค่าจำนวนนับ n ที่ไม่เกิน 100 และแสดงค่าของ F(n) แบบ  recursive (ไม่ใช้สูตร binet)

```
def fibo(n) :
  if n == 0 or n == 1 :
    return n
  else :
    return fibo(n-1) + fibo(n-2)

n = int(input())
print('f(%d) = %d' % (n, fibo(n)))
```

# **9 NumPy Array**

NumPy ย่อมาจาก Numerical Python เป็น library ที่ใช้ในการคำนวณเกี่ยวกับ ndrray โดยเฉพาะ ย่อมาจาก n-dimensional array เช่น เวกเตอร์ถือเป็น ndarray 1 มิติ, เมทริกซ์ถือเป็น ndarray 2 มิติ 

```
import numpy as np

a = np.array(1)
b = np.array([1, 2, 3, 4, 5])
c = np.array([[1, 2, 3], [4, 5, 6]])
d = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 1], [2, 1, 2]]])

print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)
```



ทุก item ใน ndarray จะต้องเป็นข้อมูลชนิดเดียวกันทั้งหมด 

ส่วนการเรียง dimension จะเรียงจากวงเล็บนอกสุดไปหาวงเล็บในสุด 

ในทุก ๆ dimension ตำแหน่งแรกจะมี index เป็น 0
